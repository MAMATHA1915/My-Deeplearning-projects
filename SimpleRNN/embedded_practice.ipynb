```python
import pandas as pd
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
#define the vocabulary size
Voc_size = 1000
```

```python
one_hot_rep = [one_hot(words,Voc_size)for words in sent]
one_hot_rep
```

```python
#word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences #making all sectences equal (add zeros for the sentence which has less words than max len of sentence)
from tensorflow.keras.models import Sequential
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_rep, padding="pre",maxlen=sent_length)
print(embedded_docs)
```

```python
#feture representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(Voc_size,dim,input_length = sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
